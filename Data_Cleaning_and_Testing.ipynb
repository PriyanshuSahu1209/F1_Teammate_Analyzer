#### Creating Summary Table for Comparison of Data
##### Function for Analyzing

```python
def analyze_driver_teammates(driver_id: int, results_df: pd.DataFrame, drivers_df: pd.DataFrame) -> pd.DataFrame:
    # Fetch driver name
    driver_row = drivers_df[drivers_df['driverId'] == driver_id]
    driver_name = f"{driver_row.iloc[0]['forename']} {driver_row.iloc[0]['surname']}" if not driver_row.empty else f"Driver ID {driver_id}"

    # Load races.csv to get season info
    races_df = pd.read_csv("datasets/races.csv")
    
    # Get all results for the target driver
    driver_results = results_df[results_df["driverId"] == driver_id].copy()
    driver_results = driver_results.merge(races_df[['raceId', 'year']], on='raceId', how='left')
    
    # Find teammates per race
    teammates = []
    for _, row in driver_results.iterrows():
        race_id = row["raceId"]
        team_id = row["constructorId"]
        
        same_team = results_df[(results_df["raceId"] == race_id) &
                               (results_df["constructorId"] == team_id) &
                               (results_df["driverId"] != driver_id)]
        
        for _, mate in same_team.iterrows():
            teammates.append({
                "year": row["year"],
                "raceId": race_id,
                "teammateId": mate["driverId"],
                "Teammate Grid": mate["grid"],
                "Teammate Pos": mate["positionOrder"],
                "Teammate Points": mate["points"],
                f"{driver_name} Grid": row["grid"],
                f"{driver_name} Pos": row["positionOrder"],
                f"{driver_name} Points": row["points"]
            })

    teammate_df = pd.DataFrame(teammates)

    # Merge with driver names
    teammate_df = teammate_df.merge(drivers_df[['driverId', 'forename', 'surname']],
                                     left_on='teammateId', right_on='driverId', how='left')
    teammate_df["Teammate Name"] = teammate_df["forename"] + " " + teammate_df["surname"]
    teammate_df.drop(columns=["driverId", "forename", "surname"], inplace=True)

    # Determine if teammate finished ahead
    teammate_df["Teammate Ahead"] = teammate_df["Teammate Pos"] < teammate_df[f"{driver_name} Pos"]

    # Group and summarize
    summary = teammate_df.groupby("Teammate Name").agg({
        "raceId": "count",
        "Teammate Ahead": "sum",
        "Teammate Points": "sum",
        f"{driver_name} Points": "sum",
        "Teammate Pos": "mean",
        f"{driver_name} Pos": "mean"
    }).rename(columns={
        "raceId": "Races Together",
        "Teammate Ahead": f"Times Teammate Beat {driver_name}",
        "Teammate Points": "Total Teammate Points",
        f"{driver_name} Points": f"Total {driver_name} Points",
        "Teammate Pos": "Avg Teammate Position",
        f"{driver_name} Pos": f"Avg {driver_name} Position"
    })

    # Calculate head-to-head % wins
    times_beat_col = f"Times Teammate Beat {driver_name}"
    summary["% Races Beaten Driver"] = 100 * summary[times_beat_col] / summary["Races Together"]

    # Round results
    return summary.round(2).sort_values("% Races Beaten Driver", ascending=False)
```

#### Visualizing The Comparison
##### Function To Render Comparison Bar Plot

```python
def plot_teammate_comparison(driver_id: int, summary_df: pd.DataFrame, drivers_df: pd.DataFrame):
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Dynamically fetch driver name
    driver_row = drivers_df[drivers_df['driverId'] == driver_id]
    driver_name = f"{driver_row.iloc[0]['forename']} {driver_row.iloc[0]['surname']}" if not driver_row.empty else f"Driver ID {driver_id}"

    sns.set(style="whitegrid")
    fig, axes = plt.subplots(1, 3, figsize=(20, 5))
    fig.suptitle(f"Teammate Comparison vs {driver_name}", fontsize=18)

    # Plot 1: % Races Beaten Driver
    sns.barplot(
        ax=axes[0],
        x=summary_df.index,
        y=summary_df['% Races Beaten Driver'],
        hue=summary_df.index,
        palette="Blues_d",
        legend=False
    )
    axes[0].set_title(f"% Races Teammate Beat {driver_name}")
    axes[0].set_ylabel("Percentage")
    axes[0].set_xlabel("Teammate Name")
    axes[0].tick_params(axis='x', rotation=45)

    # Plot 2: Average Positions
    avg_teammate_col = "Avg Teammate Position"
    avg_driver_col = f"Avg {driver_name} Position"
    avg_pos_data = summary_df[[avg_teammate_col, avg_driver_col]].reset_index()
    avg_pos_data = avg_pos_data.melt(id_vars='Teammate Name', var_name='Type', value_name='Position')
    avg_pos_data['Type'] = avg_pos_data['Type'].replace({
        avg_teammate_col: 'Avg Teammate Pos',
        avg_driver_col: f'Avg {driver_name} Pos'
    })

    sns.barplot(
        ax=axes[1],
        x='Teammate Name',
        y='Position',
        hue='Type',
        data=avg_pos_data,
        palette="pastel"
    )
    axes[1].set_title("Average Finishing Position")
    axes[1].set_xlabel("Teammate Name")
    axes[1].tick_params(axis='x', rotation=45)

    # Plot 3: Total Points Comparison
    teammate_points_col = "Total Teammate Points"
    driver_points_col = f"Total {driver_name} Points"
    points_data = summary_df[[teammate_points_col, driver_points_col]].reset_index()
    points_data = points_data.melt(id_vars='Teammate Name', var_name='Type', value_name='Points')
    points_data['Type'] = points_data['Type'].replace({
        teammate_points_col: 'Total Teammate Points',
        driver_points_col: f'Total {driver_name} Points'
    })

    sns.barplot(
        ax=axes[2],
        x='Teammate Name',
        y='Points',
        hue='Type',
        data=points_data,
        palette="muted"
    )
    axes[2].set_title("Total Points Comparison")
    axes[2].set_xlabel("Teammate Name")
    axes[2].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()
```
